```python
program_running = True   # controls whole program
while program_running:   # MAIN PROGRAM LOOP

    print("\nSTUDENT PROGRESS TRACKER SYSTEM")
    print("-----------------------------------------")

    # ----- Student Details -----
    name = input("Enter Student Name: ")
    student_class = input("Enter Class: ")
    roll = input("Enter Roll Number: ")

    # ----- Subject Marks -----
    subjects = int(input("\nHow many subjects do you want to enter? "))

    total_marks = 0
    fail_flag = False

    for i in range(1, subjects + 1):
        mark = float(input(f"Enter marks for subject {i}: "))
        total_marks += mark

        if mark < 35:
            fail_flag = True

    average = total_marks / subjects

    # ----- Grade Calculation -----
    if average >= 90:
        grade = "A"
    elif average >= 75:
        grade = "B"
    elif average >= 50:
        grade = "C"
    elif average >= 35:
        grade = "D"
    else:
        grade = "Fail"

    # ----- Pass / Fail -----
    if fail_flag or grade == "Fail":
        result = "FAIL"
    else:
        result = "PASS"

    # ----- Display Report -----
    print(" ")
    print("        STUDENT PROGRESS REPORT")
    print("-----------------------------------------")
    print("Name         :", name)
    print("Class        :", student_class)
    print("Roll Number  :", roll)
    print("Total Marks  :", total_marks)
    print("Average      :", round(average, 2))
    print("Grade        :", grade)
    print("Result       :", result)


    # ----- Menu System -----
    while True:
        print("\nWhat do you want to do next?")
        print("1. Enter new student details")
        print("2. View report again")
        print("3. Exit")

        choice = input("Enter your choice: ")

        if choice == "1":
            break   # go back to main program loop

        elif choice == "2":

            print("        STUDENT PROGRESS REPORT")
            print("-----------------------------------------")
            print("Name         :", name)
            print("Class        :", student_class)
            print("Roll Number  :", roll)
            print("Total Marks  :", total_marks)
            print("Average      :", round(average, 2))
            print("Grade        :", grade)
            print("Result       :", result)


        elif choice == "3":
            print("\nExiting Program... Goodbye!")
            program_running = False   # stop outer loop
            break   # exit menu loop

        else:
            print("Invalid choice! Please try again.")
```
